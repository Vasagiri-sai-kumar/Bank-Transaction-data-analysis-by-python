## Data collection

Key Features:

Transaction ID: Unique identifier for each transaction, ensuring traceability.

Sender Account ID: The account number of the transaction sender.

Receiver Account ID: The account number of the transaction receiver.

Transaction Amount: The monetary value involved in the transaction.

Transaction Type: The type of transaction—Transfer, Withdrawal, or Deposit.

Timestamp: The exact date and time the transaction occurred.

Transaction Status: Indicates whether the transaction was successful or failed.

Fraud Flag: A binary flag indicating whether the transaction was flagged as fraudulent.

Geolocation (Latitude/Longitude): Geographic coordinates of the transaction, helpful for spatial analysis.

Device Used: The type of device used for conducting the transaction (Mobile or Desktop).

Network Slice ID: Identifies the 6G network slice used during the transaction.

Latency (ms): The delay (in milliseconds) experienced during the transaction.

Slice Bandwidth (Mbps): The bandwidth available in the network slice used for the transaction.

PIN Code: A four-digit security code used for the transaction (masked for privacy).

## Data Visualization

```python
# import the matplotlib 
%matplotlib inline 
import matplotlib.pyplot as plt

# let data add 
data_tra.plot.hist(subplots=True);

# without using seaborn
plt.show()
```

```python
data_tra.head()
```

```python
# Available styles in matplotlib 
plt.style.available
```

```python
# using seaborn style in plt 
plt.style.use( 'seaborn-v0_8-whitegrid')
plt.style.use('seaborn-v0_8-colorblind')
plt.style.use('seaborn-v0_8')

#create two graph 
fig,(ax1,ax2) = plt.subplots(nrows=2, 
                            ncols=1, 
                            figsize=(10,10))

# create transaction amount based on transaction type bar graph
bar1 = ax1.bar(data_tra['TransactionType'],data_tra['Transaction_Amount'])

# ax1 Add x and y axis label and title name
ax1.set(title='Transaction Amount based on Transaction Type', 
        xlabel='Transaction Type', 
        ylabel='Transaction Amount')

# create scatter plot 
scatt = ax2.scatter(x=data_tra['Transaction_Amount'], 
                    y=data_tra['Latency (ms)'], 
                    c=data_tra['Fraud Flag'], 
                    cmap='winter')

# ax2 add legend 
ax2.legend(*scatt.legend_elements(), title='Fraud Flag')

# ax2 add axis label name(X&Y) and title of graph 
ax2.set(title='Scatter plot for transaction', 
        xlabel='Transaction Amount', 
        ylabel='Latency (ms)')

# ax2 add axis mean line 
ax2.axhline(data_tra['Latency (ms)'].mean(), 
            linestyle = '--', 
            color='black')

# ax2 add limit for x&y axis
ax2.set_xlim([0,1500])
ax2.set_ylim([2.5,22.5])

# Title subplots 
fig.suptitle('TRANSACTION DATA ANALYSIS', 
             fontsize=20, 
             fontweight='bold')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,6))

# create historgram based on Timestamp
hist1 = ax.hist(data_tra['Timestamp'])

# add label (x and y) and title name 
ax.set(title='Timestamp Flow analysis', 
       xlabel='Date of Transaction', 
       ylabel='Time stamp Flow')


plt.show()
```
